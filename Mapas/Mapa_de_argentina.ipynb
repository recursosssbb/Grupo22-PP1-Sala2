```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
from matplotlib import style
import seaborn as sns
import folium
from folium.plugins import MarkerCluster
```

```python
# Lee el archivo CSV en un DataFrame
data = pd.read_csv('/dataframe_completo.csv')

# Obtiene un listado de todas las columnas
columnas = data.columns

# Imprime las columnas
print("Listado de todas las columnas:")
for columna in columnas:
    print(columna)
```

MAPA 1

```python
import folium

# Crear un mapa centrado en Argentina
mapa = folium.Map(location=[-34.6118, -58.4173], zoom_start=5, tiles='CartoDB Positron')  # Tiles='CartoDB Positron' es un ejemplo de mapa base ligero

# Iterar a través de las filas de tu conjunto de datos
for index, row in data.iterrows():
    # Obtener la latitud y longitud de cada localidad
    latitud = row['Latitud']
    longitud = row['Longitud']

    # Obtener la cantidad de usuarios (ajusta la columna según tu conjunto de datos)
    usuarios = row['Total_2022']

    # Agregar un marcador para cada localidad con la cantidad de usuarios como etiqueta
    folium.CircleMarker(location=[latitud, longitud], radius=1, color='blue', fill=True,
                        fill_color='blue', fill_opacity=0.4, popup=f'Usuarios: {usuarios}').add_to(mapa)

# Guardar el mapa en un archivo HTML
mapa.save('mapa_usuarios.html')
```

```python
# Crear un mapa centrado en Argentina
mapa = folium.Map(location=[-34.6118, -58.4173], zoom_start=5, tiles='CartoDB Positron')  # Tiles='CartoDB Positron' es un ejemplo de mapa base ligero

# Crear un grupo de marcadores para agrupar los marcadores
marker_cluster = MarkerCluster().add_to(mapa)

# Iterar a través de las filas de tu conjunto de datos
for index, row in data.iterrows():
    # Obtener la latitud y longitud de cada localidad
    latitud = row['Latitud']
    longitud = row['Longitud']

    # Obtener la cantidad de usuarios (ajusta la columna según tu conjunto de datos)
    usuarios = row['Total_2022']

    # Crear un marcador y agregarlo al grupo de marcadores
    folium.CircleMarker(location=[latitud, longitud], radius=2, color='blue', fill=True,
                        fill_color='blue', fill_opacity=0.6, popup=f'Usuarios: {usuarios}').add_to(marker_cluster)

# Guardar el mapa en un archivo HTML
mapa.save('mapa_usuarios_con_agrupamiento.html')

```
